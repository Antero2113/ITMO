```python
import soundfile as sf
import numpy as np
import librosa.feature as lf
import matplotlib.pyplot as plt

# Запись двух аудиофайлов
file1 = 'Я.wav'
file2 = 'Салатик.wav'
phrase1 = 'Я обещаю сходить с тобой в театр'
phrase2 = 'Я обещаю сходить с тобой в театр'

# Загрузка аудиофайлов
data1, sr1 = sf.read(file1)
data2, sr2 = sf.read(file2)

sf.write(file1, data1, sr1)
sf.write(file2, data2, sr2)

# Параметры извлечения MFCC
n_mfcc = 13

# Извлечение характеристик речи
mfcc1 = lf.mfcc(y=data1, sr=sr1, n_mfcc=n_mfcc)
mfcc2 = lf.mfcc(y=data2, sr=sr2, n_mfcc=n_mfcc)
```

```python
# Спектрограммы
spectrogram1 = lf.melspectrogram(y=data1, sr=sr1)
spectrogram2 = lf.melspectrogram(y=data2, sr=sr2)

# Спектральный центроид
spectral_centroid1 = lf.spectral_centroid(y=data1, sr=sr1)
spectral_centroid2 = lf.spectral_centroid(y=data2, sr=sr2)

# Спектральный контраст
spectral_contrast1 = lf.spectral_contrast(y=data1, sr=sr1)
spectral_contrast2 = lf.spectral_contrast(y=data2, sr=sr2)

# Спектральный спад
spectral_rolloff1 = lf.spectral_rolloff(y=data1, sr=sr1)
spectral_rolloff2 = lf.spectral_rolloff(y=data2, sr=sr2)

# Спектральная ширина
spectral_bandwidth1 = lf.spectral_bandwidth(y=data1, sr=sr1)
spectral_bandwidth2 = lf.spectral_bandwidth(y=data2, sr=sr2)

# Скорость пересечения нуля
zero_crossing_rate1 = lf.zero_crossing_rate(y=data1)
zero_crossing_rate2 = lf.zero_crossing_rate(y=data2)
```

```python
# Сравнение значений
print("Сравнение характеристик речи:")

print("MFCC:")
print(np.array_equal(mfcc1, mfcc2))

print("Спектрограммы:")
print(np.array_equal(spectrogram1, spectrogram2))

print("Спектральный центроид:")
print(np.array_equal(spectral_centroid1, spectral_centroid2))

print("Спектральный контраст:")
print(np.array_equal(spectral_contrast1, spectral_contrast2))

print("Спектральный спад:")
print(np.array_equal(spectral_rolloff1, spectral_rolloff2))

print("Спектральная ширина:")
print(np.array_equal(spectral_bandwidth1, spectral_bandwidth2))

print("Скорость пересечения нуля:")
print(np.array_equal(zero_crossing_rate1, zero_crossing_rate2))

# Визуализация
plt.figure()
plt.subplot(2, 1, 1)
plt.imshow(mfcc1, interpolation='nearest')
plt.xlabel('Время')
plt.ylabel('Коэффициенты MFCC')
plt.title('MFCC для файла {}'.format(file1))

plt.subplot(2, 1, 2)
plt.imshow(mfcc2, interpolation='nearest')
plt.xlabel('Время')
plt.ylabel('Коэффициенты MFCC')
plt.title('MFCC для файла {}'.format(file2))

plt.show()
```

```python
# Расчет расстояний между характеристиками
mfcc1_1 = mfcc1[:, :225]
distance_mfcc = np.linalg.norm(mfcc1_1 - mfcc2)
spectrogram1_1 = spectrogram1[:, :225]
distance_spectrogram = np.linalg.norm(spectrogram1_1 - spectrogram2)
spectral_centroid1_1 = spectral_centroid1[:, :225]
distance_spectral_centroid = np.linalg.norm(spectral_centroid1_1 - spectral_centroid2)
spectral_contrast1_1 = spectral_contrast1[:, :225]
distance_spectral_contrast = np.linalg.norm(spectral_contrast1_1 - spectral_contrast2)
spectral_rolloff1_1 = spectral_rolloff1[:, :225]
distance_spectral_rolloff = np.linalg.norm(spectral_rolloff1_1 - spectral_rolloff2)
spectral_bandwidth1_1 = spectral_bandwidth1[:, :225]
distance_spectral_bandwidth = np.linalg.norm(spectral_bandwidth1_1 - spectral_bandwidth2)
zero_crossing_rate1_1 = zero_crossing_rate1[:, :225]
distance_zero_crossing_rate = np.linalg.norm(zero_crossing_rate1_1 - zero_crossing_rate2)

# Оценка эффективности
print("Эффективность характеристик речи в идентификации человека по голосу:")
print("MFCC:", distance_mfcc)
print("Спектрограмма:", distance_spectrogram)
print("Спектральный центроид:", distance_spectral_centroid)
print("Спектральный контраст:", distance_spectral_contrast)
print("Спектральный спад:", distance_spectral_rolloff)
print("Спектральная ширина:", distance_spectral_bandwidth)
print("Скорость пересечения нуля:", distance_zero_crossing_rate)
```
